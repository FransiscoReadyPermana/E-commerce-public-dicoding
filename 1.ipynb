# Data Wrangling

## Gathering Data

### Order Dataset

```python
order_dataset = pd.read_csv('../Data/orders_dataset.csv')
order_dataset.head()
```

### Customer Dataset

```python
customer_dataset = pd.read_csv('../Data/customers_dataset.csv')
customer_dataset.head()
```

### Order Reviews Dataset

```python
order_reviews_dataset = pd.read_csv('../Data/order_reviews_dataset.csv')
order_reviews_dataset.head()
```

### Order Payments Dataset

```python
order_payments_dataset = pd.read_csv('../Data/order_payments_dataset.csv')
order_payments_dataset.head()
```

### Order Items Dataset

```python
order_items_dataset = pd.read_csv('../Data/order_items_dataset.csv')
order_items_dataset.head()
```

### Product Dataset

```python
product_dataset = pd.read_csv('../Data/products_dataset.csv')
product_dataset.head()
```

### Product Category Name Translation

```python
product_category_name_translation = pd.read_csv('../Data/product_category_name_translation.csv')
product_category_name_translation.head()
```

### Membuat Tabel Product
(product_dataset X product_category_name_translation)

trus ambil product_id sama product_category_name_english

```python
product = pd.merge(product_dataset, product_category_name_translation, on='product_category_name')
product = product[['product_id', 'product_category_name_english']]
product.head()
```

### Membuat Tabel Order Product
(order_items_dataset X product)

trus ambil order_id, product_category_name_english, freight_value, price, shipping_limit_date

```python
order_items = pd.merge(order_items_dataset, product, on='product_id')
order_items = order_items[['order_id', 'product_category_name_english', 'freight_value', 'price', 'shipping_limit_date']]
order_items.head()
```

### Membuat Tabel df

#### order_dataset X customer_dataset

```python
order_dataset
```

```python
df = pd.merge(order_dataset, customer_dataset[['customer_id', 'customer_city']], on='customer_id', how='left')
df
```

#### df X order_reviews_dataset

```python
df = pd.merge(df, order_reviews_dataset[['review_score', 'order_id']], left_on='order_id', right_on='order_id' ,how='left')
df
```

#### df X Order Payment Dataset

```python
df = pd.merge(df, order_payments_dataset, on='order_id')
df
```

#### df X order_items

## Assesing Data

nah untuk sementara, berikut adalah data serta nomenklatur dari setiap kolomnya dari data yang akan saya analisis.

<b> Nomenklatur Kolom </b>
- order_id : id setiap ordernya
- customer_id : id dari setiap customernya
- order_status : Status dari setiap ordernya
- order_purchase_timestamp : waktu dan tanggal ketika suatu pesanan atau order dibuat atau diproses
- order_approved_at : waktu dan tanggal ketika suatu pesanan atau order dalam suatu sistem e-commerce disetujui atau diverifikasi
- Order_delivered_carrier_date : waktu ketika pesanan diserahkan kepada penyedia pengiriman.
- Order_delivered_customer_date : waktu ketika pesanan tiba di tangan pelanggan.
- Order_estimated_delivery_date : tanggal perkiraan pengiriman pesanan.
- Customer_city : kota tempat tinggal pelanggan.
- Review_score : nilai atau skor yang diberikan oleh pelanggan sebagai umpan balik atas pengalaman berbelanja mereka.
- Payment_sequential : urutan pembayaran dalam satu pesanan.
- Payment_type : jenis metode pembayaran yang digunakan oleh pelanggan.
- Payment_installments : jumlah pembayaran angsuran yang dibuat untuk pesanan.
- Payment_value : total nilai pembayaran dalam satu pesanan.
- Product_category_name_english : kategori produk dalam bahasa Inggris.
- Freight_value : biaya pengiriman atau ongkir.
- Price : harga produk.
- Shipping_limit_date : batas waktu untuk pengiriman pesanan.

```python
# make plot to see the distribution of the data
fig, ax = plt.subplots(2, 2, figsize=(20, 20))
fig.suptitle("Distribusi Data", fontsize=20, y=0.9)
sns.histplot(df["review_score"], ax=ax[0,0], fill=False, stat="density").set_xlabel('review_score', fontsize=14)
sns.kdeplot(df["review_score"], ax=ax[0,0], color="red").set_ylabel('Density', fontsize=14)

sns.histplot(df["freight_value"], ax=ax[0,1], fill=False, stat="density").set_xlabel('freight_value', fontsize=14)
sns.kdeplot(df["freight_value"], color="red", ax=ax[0,1]).set_ylabel('Density', fontsize=14)

sns.histplot(df["price"], ax=ax[1,0], fill=False, stat="density").set_xlabel('price', fontsize=14)
sns.kdeplot(df["price"], color="red", ax=ax[1,0]).set_ylabel('Density', fontsize=14)
```
