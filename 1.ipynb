## Gathering Data

### Order Dataset

```python
order_dataset = pd.read_csv('../Data/orders_dataset.csv', parse_dates=['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date'])
order_dataset.head()
```

### Order Reviews Dataset

```python
order_reviews_dataset = pd.read_csv('../Data/order_reviews_dataset.csv', parse_dates=['review_creation_date', 'review_answer_timestamp'])
order_reviews_dataset.head()
```

### Order Payments Dataset

```python
order_payments_dataset = pd.read_csv('../Data/order_payments_dataset.csv')
order_payments_dataset.head()
```

### Order Items Dataset

```python
order_items_dataset = pd.read_csv('../Data/order_items_dataset.csv', parse_dates=['shipping_limit_date'])
order_items_dataset.head()
```

### Product Dataset

```python
product_dataset = pd.read_csv('../Data/products_dataset.csv')
product_dataset.head()
```

### Product Category Name Translation

```python
product_category_name_translation = pd.read_csv('../Data/product_category_name_translation.csv')
product_category_name_translation.head()
```

### Membuat Tabel Product
(product_dataset X product_category_name_translation)

trus ambil product_id sama product_category_name_english

```python
product = pd.merge(product_dataset, product_category_name_translation, on='product_category_name')
product = product[['product_id', 'product_category_name_english']]
product.head()
```

### Membuat Tabel Order Product
(order_items_dataset X product)

trus ambil order_id, product_category_name_english, freight_value, price, shipping_limit_date

```python
order_items = pd.merge(order_items_dataset, product, on='product_id')
order_items = order_items[['order_id', 'product_category_name_english', 'freight_value', 'price']]
order_items.head()
```

karena saya menemukan bahwa hanya ada 7 data yang bernilai canceled, jadi saya berpikir untuk mengambil data dengan nilai delivered dan menghapus kolom order status

karena datanya banyak yang double karena hasil join dengan payment sequential, jadi saya mau menyatukannya lagi sehingga tidak ada data yang double lagi. saya melakukannya dengan menggroup by datanya lalu meng aggregatenya dengan ketentuan sebagai berikut

```python
df = df.groupby('order_id').agg({
  'customer_id': 'first',
  'order_purchase_timestamp': 'first',
  'order_approved_at': 'first',
  'order_delivered_carrier_date': 'first',
  'order_delivered_customer_date': 'first',
  'order_estimated_delivery_date': 'first',
  'customer_city': 'first', 
  'review_score': 'first', 
  'product_category_name_english': 'first', 
  'payment_type': 'max',
  'payment_sequential': 'max',
  'payment_value': 'sum',
  'price': 'sum', 
  'freight_value': 'sum'})
df.reset_index(inplace=True)
df
```
