## 7.  Bagaimana analisis RFM nya

### Mencari Jawabannya

#### Recency

```python
df
```

```python
latest_date = df['order_purchase_timestamp'].max()
recency = df.groupby(by='customer_id', as_index=False)['order_purchase_timestamp'].max()
recency['Recency'] = recency['order_purchase_timestamp'].apply(lambda x: (latest_date - x).days)
recency
```

#### Frequency

```python
frequency = df.groupby(by=['customer_id'], as_index=False)['order_id'].count()
frequency.columns = ['customer_id', 'Frequency']
frequency
```

#### Monetary

```python
monetary = df.groupby(by=['customer_id'], as_index=False)['price'].sum()
monetary.columns = ['customer_id', 'Monetary']
monetary
```

#### RFM

```python
df_rfm = recency.merge(frequency, on='customer_id')
df_rfm = df_rfm.merge(monetary, on='customer_id')
df_rfm
```

```python
df_rfm['R'] = pd.qcut(df_rfm['Recency'], 4, ['4', '3', '2', '1'])
df_rfm['F'] = pd.qcut(df_rfm['Frequency'].rank(method='first'), 4, ['1', '2', '3', '4'])
df_rfm['M'] = pd.qcut(df_rfm['Monetary'], 4, ['1', '2', '3', '4'])
df_rfm
```

```python
df_rfm['RFM Score'] = df_rfm['R'].astype('int') + df_rfm['F'].astype('int') + df_rfm['M'].astype('int')
df_rfm
```

```python
df_rfm.sort_values(by='RFM Score', ascending=False)
```

```python
df_rfm['RFM Score'].value_counts().sort_values(ascending=False)
```

```python
rfm_level = ['Bronze', 'Silver', 'Gold', 'Platinum']

def determine_level(x):
    if x >= 11:
        return 'Platinum'
    elif x >= 7:
        return 'Gold'
    else:
        return 'Silver'

df_rfm['RFM Level'] = df_rfm['RFM Score'].apply(determine_level)
df_rfm
```

```python
# Meresample data ke bulanan dan menghitung jumlah pesanan
RFM = df_rfm['RFM Level'].value_counts().sort_values(ascending=False)
RFM.columns = ['RFM Level', 'Count']

# Membuat plot interaktif dengan Plotly Express

fig = px.pie(RFM, values=RFM.values, names=RFM.index, title='RFM Level',
              labels={'index': 'RFM Level', 'value': 'Count'}, height=600, color_discrete_sequence=['#C76249', '#F3AE4E', '#F9E5A3', '#F3E1C1'])


# Menambahkan mode hover untuk menampilkan detail saat dihover
fig.update_traces(hovertemplate='RFM Level: %{label}<br>Count: %{value}', hoverlabel=dict(bgcolor='antiquewhite', font_size=12, font_family='Rockwell'), name='')
fig.update_layout(showlegend=False)

fig.show()
```
